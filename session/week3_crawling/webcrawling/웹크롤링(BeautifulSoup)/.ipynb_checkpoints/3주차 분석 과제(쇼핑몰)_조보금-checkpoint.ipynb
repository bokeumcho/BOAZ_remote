```python
page = 1
url1='http://deal.11st.co.kr/product/SellerProductDetail.tmall?method=getProductReviewList&prdNo=87595509&page='
url2='&pageTypCd=first&reviewDispYn=Y&isPreview=false&reviewOptDispYn=n&optSearchBtnAndGraphLayer=n&reviewBottomBtn=Y&openDetailContents=Y&pageSize=10&isIgnoreAuth=false&lctgrNo=1001369&leafCtgrNo=0&groupProductNo=0&groupFirstViewPrdNo=0&selNo=16674487#this'

page_num=[]
review=[]
star1=[]

while page <= 15:
    url = url1 + str(page) + url2
    
    html=rq.get(url).text
    soup = bs4.BeautifulSoup(html,'lxml')
    soup.prettify()    
    
    bokeums=soup.find_all("span",{"class":"summ_conts"})
    for bokeum in bokeums:
        review.append(bokeum.text.strip())
        page_num.append(page)

    star=soup.find_all("div",{"class":"selr_wrap"})
    for n in star:
        star_real=re.findall('\d+',n.text.strip())
        try :
            star1.append(star_real[1])
        except : 
            break
    page=page+1    
```

```python
del star1[-5:-1]
len(star1)
```

```python
result=pd.DataFrame({'':page_num,'':star1,'':review})
result
```
