```python
import requests as rq
import pandas as pd
from pandas import DataFrame
import bs4
from bs4 import BeautifulSoup
import re
```

```python
url = "http://deal.11st.co.kr/product/SellerProductDetail.tmall?method=getProductReviewList&prdNo=87595509&page=1&pageTypCd=first&reviewDispYn=Y&isPreview=false&reviewOptDispYn=n&optSearchBtnAndGraphLayer=n&reviewBottomBtn=Y&openDetailContents=Y&pageSize=10&isIgnoreAuth=false&lctgrNo=1001369&leafCtgrNo=0&groupProductNo=0&groupFirstViewPrdNo=0&selNo=16674487#this"
res = rq.get(url) 
print(res.content) #바이너리 형태로 HTML을 가져온다. res.text로도 가져올 수 있다. 
```

```python
import bs4
from bs4 import BeautifulSoup
html = rq.get(url).text
soup = bs4.BeautifulSoup(html,'lxml') # 파서란 원시코드인 순수 문자열 객체를 해석할 수 있도록 분석하는것. 
                                      # 속도가 빠르기 때문에 lxml을 보통 쓴다.
print(soup.prettify())
```

```python
review = []
bokeums=soup.find_all("span",{"class":"summ_conts"})

for bokeum in bokeums:
    review.append(bokeum.text.strip())

print(review)
```

```python
star1=[]

import re
for n in star:
    star_real=re.findall('\d+',n.text.strip())
    star1.append(star_real[1])
star1
```

```python
page = 1
url1='http://deal.11st.co.kr/product/SellerProductDetail.tmall?method=getProductReviewList&prdNo=87595509&page='
url2='&pageTypCd=first&reviewDispYn=Y&isPreview=false&reviewOptDispYn=n&optSearchBtnAndGraphLayer=n&reviewBottomBtn=Y&openDetailContents=Y&pageSize=10&isIgnoreAuth=false&lctgrNo=1001369&leafCtgrNo=0&groupProductNo=0&groupFirstViewPrdNo=0&selNo=16674487#this'

page_num=[]
review=[]
star1=[]

while page <= 15:
    url = url1 + str(page) + url2
    
    html=rq.get(url).text
    soup = bs4.BeautifulSoup(html,'lxml')
    soup.prettify()    
    
    bokeums=soup.find_all("span",{"class":"summ_conts"})
    for bokeum in bokeums:
        review.append(bokeum.text.strip())
        page_num.append(page)

    star=soup.find_all("div",{"class":"selr_wrap"})
    for n in star:
        star_real=re.findall('\d+',n.text.strip())
        try :
            star1.append(star_real[1])
        except : 
            break
    page=page+1    
```

```python
del star1[-5:-1]
len(star1)
```

```python
result=pd.DataFrame({'페이지':page_num,'별점':star1,'리뷰':review})
result
```
